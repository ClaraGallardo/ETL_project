# Extraccion de datos Tripadvisor.es:

Comenzamos definiedo la funcion de extraccion.

Definimos las url de las que queremos extrar la informacion ya seleccionada y la guardamos en una lista, con la que luego nos haremos nuestro DF.

```python
def extraer(url):

    global tabla

    driver = webdriver.Chrome()
    driver.get(url)

    time.sleep(2)
    
    # COOKIES

    try:

        aceptar = driver.find_element(By.XPATH, '//*[@id="onetrust-accept-btn-handler"]')
        aceptar.click()

    except:

        print('Ya')

    while True: # ejecutamos hasta que nos haya descargado todo ü§ûü§û

        time.sleep(5)

        tarjetas = []

        for e in range(len(driver.find_elements(By.TAG_NAME, 'header'))):
            box = driver.find_elements(By.TAG_NAME, 'header')[e].text.split('\n')
            tarjetas.append(box)
        
        tabla += tarjetas

        # HACEMOS SCROLL:

        time.sleep(5)

        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        # CLICK SIGUIENTE:

        try:

            time.sleep(5)

            aceptar = driver.find_element(By.XPATH, '//*[@id="lithium-root"]/main/div/div[4]/div[6]/div/div[1]/div[2]/div/a')
            aceptar.click()

        except:
            try:

                aceptar = driver.find_element(By.CSS_SELECTOR, '#lithium-root > main > div > div.SuxDz > div:nth-child(6) > div > div.OvVFl.j > div.xkSty > div > a')
                aceptar.click()

            except:
                print('No se encontr√≥ el bot√≥n "siguiente"')
                break  # Salir del bucle si no se encuentra el bot√≥n "siguiente"
    
    
```
