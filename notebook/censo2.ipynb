```python
import requests
import pandas as pd

url = 'https://datos.madrid.es/egob/catalogo/209548-485-censo-locales-historico.json'  

response = requests.get(url)

if response.status_code == 200:
    
    data = response.json()
    df = pd.DataFrame(data)

else:
    print("No se pudo descargar el archivo JSON.")
```

```python
data=df.copy()
data.head()
```
